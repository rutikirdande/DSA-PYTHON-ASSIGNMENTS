```python
# <aside>
# ðŸ’¡ Question-1

# You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

# Input:

#         10

#        /   \

#      2      7

#    /   \

#  8      4

# Output:

#         8

#       /   \

#     4     10

#   /   \

# 2      7

# </aside>
class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def convertToBST(root):
    # Perform in-order traversal and store node values in a list
    def inorderTraversal(node, nodes):
        if node:
            inorderTraversal(node.left, nodes)
            nodes.append(node)
            inorderTraversal(node.right, nodes)

    nodes = []
    inorderTraversal(root, nodes)

    # Sort the node values
    nodeValues = [node.val for node in nodes]
    nodeValues.sort()

    # Update the node values using the sorted list
    def updateNodeValues(node, index):
        if node:
            updateNodeValues(node.left, index)
            node.val = nodeValues[index[0]]
            index[0] += 1
            updateNodeValues(node.right, index)

    index = [0]
    updateNodeValues(root, index)

    return root

root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

convertedRoot = convertToBST(root)
print(convertedRoot.val)  # Output: 8
print(convertedRoot.left.val)  # Output: 4
print(convertedRoot.right.val)  # Output: 10
print(convertedRoot.left.left.val)  # Output: 2
print(convertedRoot.left.right.val)  # Output: 7
```

```python
# <aside>
# ðŸ’¡ Question-2:

# Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

# Example:

# Consider the following BST:

# ![1.png](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/f2455039-7e12-43fc-a7d3-b5be24931c1c/1.png)

# **Input-1:**

# n = 9

# values = [8, 3, 1, 6, 4, 7, 10, 14,13]

# node-1 = 6

# node-2 = 14

# **Output-1:**

# The distance between the two keys = 4

# **Input-2:**

# n = 9

# values = [8, 3, 1, 6, 4, 7, 10, 14,13]

# node-1 = 3

# node-2 = 4

# **Output-2:**

# The distance between the two keys = 2

# </aside>

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def findDistance(root, node1, node2):
    # Find the Lowest Common Ancestor (LCA) of the two nodes
    def findLCA(root, node1, node2):
        if not root:
            return None

        if root.val > node1.val and root.val > node2.val:
            return findLCA(root.left, node1, node2)
        elif root.val < node1.val and root.val < node2.val:
            return findLCA(root.right, node1, node2)
        else:
            return root

    # Calculate the distance from the LCA to a given node
    def findDistanceToNode(root, node, distance):
        if not root:
            return 0

        if root.val == node.val:
            return distance

        if root.val > node.val:
            return findDistanceToNode(root.left, node, distance + 1)
        else:
            return findDistanceToNode(root.right, node, distance + 1)

    # Find the LCA
    lca = findLCA(root, node1, node2)

    # Calculate the distance from the LCA to each node
    distance1 = findDistanceToNode(lca, node1, 0)
    distance2 = findDistanceToNode(lca, node2, 0)

    # Calculate the distance between the two nodes
    distance = distance1 + distance2

    return distance


# Example 1
root = TreeNode(8)
root.left = TreeNode(3)
root.right = TreeNode(10)
root.left.left = TreeNode(1)
root.left.right = TreeNode(6)
root.left.right.left = TreeNode(4)
root.left.right.right = TreeNode(7)
root.right.right = TreeNode(14)
root.right.right.left = TreeNode(13)

node1 = TreeNode(6)
node2 = TreeNode(14)

distance = findDistance(root, node1, node2)
print("The distance between the two keys:", distance)  # Output: 4

# Example 2
node1 = TreeNode(3)
node2 = TreeNode(4)

distance = findDistance(root, node1, node2)
print("The distance between the two keys:", distance)  # Output: 2
```

```python
# <aside>
# ðŸ’¡ Question-3:

# Write a program to convert a binary tree to a doubly linked list.

# Input:

#         10

#        /   \

#      5     20

#            /   \

#         30     35

# Output:

# 5 10 30 20 35

# </aside>

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, value):
        self.val = value
        self.prev = None
        self.next = None

def convertToDoublyLinkedList(root):
    if not root:
        return None

    # Helper function to perform in-order traversal
    def inorderTraversal(node):
        nonlocal prev, head

        if not node:
            return

        inorderTraversal(node.left)

        # Create a new doubly linked list node
        new_node = DoublyLinkedListNode(node.val)

        if not prev:
            head = new_node
        else:
            prev.next = new_node
            new_node.prev = prev

        prev = new_node

        inorderTraversal(node.right)

    prev = None
    head = None

    # Perform in-order traversal to convert the binary tree to doubly linked list
    inorderTraversal(root)

    return head


# Create the binary tree
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

# Convert the binary tree to doubly linked list
head = convertToDoublyLinkedList(root)

# Print the doubly linked list
current = head
while current:
    print(current.val, end=" ")
    current = current.next

# Output: 5 10 30 20 35
```

```python
# <aside>
# ðŸ’¡ Question-4:

# Write a program to connect nodes at the same level.

# Input:

#         1

#       /   \

#     2      3

#   /   \   /   \

# 4     5 6    7

# Output:

# 1 â†’ -1

# 2 â†’ 3

# 3 â†’ -1

# 4 â†’ 5

# 5 â†’ 6

# 6 â†’ 7

# 7 â†’ -1

# </aside>

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        self.next = None

def connectNodesAtSameLevel(root):
    if not root:
        return

    # Perform level-order traversal using a queue
    queue = [root]

    while queue:
        size = len(queue)

        # Connect nodes at the current level
        for i in range(size):
            node = queue.pop(0)

            if i < size - 1:
                node.next = queue[0]

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    return root
```
