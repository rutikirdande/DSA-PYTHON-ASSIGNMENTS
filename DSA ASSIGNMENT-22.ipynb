```python
# <aside>
# ðŸ’¡ Question-1:

# Given a Binary Tree (Bt), convert it to a Doubly Linked List(DLL). The left and right pointers in nodes are to be used as previous and next pointers respectively in converted DLL. The order of nodes in DLL must be the same as in Inorder for the given Binary Tree. The first node of Inorder traversal (leftmost node in BT) must be the head node of the DLL.

# </aside>

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, value):
        self.val = value
        self.prev = None
        self.next = None

def convertBinaryTreeToDLL(root):
    if not root:
        return None

    # Helper function for inorder traversal
    def inorder(node, prev):
        nonlocal head, tail

        if not node:
            return prev

        # Recursively convert the left subtree
        prev = inorder(node.left, prev)

        # Create a new DLL node for the current tree node
        new_node = DoublyLinkedListNode(node.val)

        # Connect the previous node to the current node
        if prev:
            prev.next = new_node
            new_node.prev = prev
        else:
            # The first node encountered in inorder traversal will be the head of the DLL
            head = new_node

        # Update the tail of the DLL
        tail = new_node

        # Recursively convert the right subtree
        prev = inorder(node.right, new_node)

        return prev

    head = None
    tail = None

    # Start the inorder traversal from the root
    inorder(root, None)

    # Return the head of the DLL
    return head


# Create the binary tree
root = TreeNode(4)
root.left = TreeNode(2)
root.right = TreeNode(5)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)

# Convert the binary tree to a DLL
dll_head = convertBinaryTreeToDLL(root)

# Print the DLL in both forward and backward directions
current = dll_head
forward = "Forward: "
backward = "Backward: "
while current:
    forward += str(current.val) + " "
    current = current.next

current = dll_head
while current.next:
    current = current.next

while current:
    backward += str(current.val) + " "
    current = current.prev

print(forward)
print(backward)

# Output:
# Forward: 1 2 3 4 5
# Backward: 5 4 3 2 1
```

```python
# <aside>
# ðŸ’¡ Question-3:

# Given a binary tree, print all its root-to-leaf paths without using recursion. For example, consider the following Binary Tree.

# Input:

#         6
#      /    \
#     3      5
#   /   \     \
#  2     5     4
#      /   \
#     7     4

# Output:

# There are 4 leaves, hence 4 root to leaf paths -
#   6->3->2
#   6->3->5->7
#   6->3->5->4
#   6->5>4

# </aside>

class TreeNode:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None

def printRootToLeafPaths(root):
    if not root:
        return

    stack = []
    path_stack = []

    stack.append(root)
    path_stack.append(str(root.val))

    while stack:
        node = stack.pop()
        path = path_stack.pop()

        if not node.left and not node.right:
            # Leaf node reached, print the path
            print(path)
        
        if node.right:
            stack.append(node.right)
            path_stack.append(path + "->" + str(node.right.val))
        
        if node.left:
            stack.append(node.left)
            path_stack.append(path + "->" + str(node.left.val))

            
# Create the binary tree
root = TreeNode(6)
root.left = TreeNode(3)
root.right = TreeNode(5)
root.left.left = TreeNode(2)
root.left.right = TreeNode(5)
root.right.right = TreeNode(4)
root.left.right.left = TreeNode(7)
root.left.right.right = TreeNode(4)

# Print the root-to-leaf paths
printRootToLeafPaths(root)

# Output:
# 6->3->2
# 6->3->5->7
# 6->3->5->4
# 6->5->4
```

```python
# <aside>
# ðŸ’¡ Question-4:

# GivenÂ Preorder, Inorder and PostorderÂ traversals of some tree. Write a program to check if they all are of the same tree.

# **Examples:**

# Input : 

#         Inorder -> 4 2 5 1 3
#         Preorder -> 1 2 4 5 3
#         Postorder -> 4 5 2 3 1
# Output : 

# Yes
# Explanation : 

# All of the above three traversals are of
# the same tree 

#                            1
#                          /   \
#                         2     3
#                       /   \
#                      4     5

# Input : 

#         Inorder -> 4 2 5 1 3
#         Preorder -> 1 5 4 2 3
#         Postorder -> 4 1 2 3 5
# Output : 

# No

# </aside>

def isSameTree(preorder, inorder, postorder):
    if len(preorder) == 0:
        return True
    
    if len(preorder) == 1:
        return (preorder == inorder) and (preorder == postorder)
    
    root = preorder[0]
    root_idx = inorder.index(root)
    
    left_inorder = inorder[:root_idx]
    right_inorder = inorder[root_idx+1:]
    
    left_preorder = preorder[1:root_idx+1]
    right_preorder = preorder[root_idx+1:]
    
    left_postorder = postorder[:root_idx]
    right_postorder = postorder[root_idx:len(postorder)-1]
    
    return isSameTree(left_preorder, left_inorder, left_postorder) and isSameTree(right_preorder, right_inorder, right_postorder)

# Example 1
inorder1 = [4, 2, 5, 1, 3]
preorder1 = [1, 2, 4, 5, 3]
postorder1 = [4, 5, 2, 3, 1]
print(isSameTree(preorder1, inorder1, postorder1))  # Output: True

# Example 2
inorder2 = [4, 2, 5, 1, 3]
preorder2 = [1, 5, 4, 2, 3]
postorder2 = [4, 1, 2, 3, 5]
print(isSameTree(preorder2, inorder2, postorder2))  # Output: False
```
