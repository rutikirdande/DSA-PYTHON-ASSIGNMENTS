```python
# <aside>
# ğŸ’¡ . **Merge k Sorted Lists**

# You are given an array ofÂ `k`Â linked-listsÂ `lists`, each linked-list is sorted in ascending order.

# *Merge all the linked-lists into one sorted linked-list and return it.*

# **Example 1:**

# </aside>
from heapq import heappush, heappop

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeKLists(lists):
    # Create a min heap
    min_heap = []
    
    # Add the first element of each list to the min heap
    for lst in lists:
        if lst:
            heappush(min_heap, (lst.val, lst))

    # Create a dummy node and a pointer to the dummy node
    dummy = ListNode()
    curr = dummy

    # Merge the lists
    while min_heap:
        val, node = heappop(min_heap)
        curr.next = node
        curr = curr.next
        if node.next:
            heappush(min_heap, (node.next.val, node.next))

    return dummy.next

```

```python
# <aside>
# ğŸ’¡ 2. **Count of Smaller Numbers After Self**

# Given an integer arrayÂ `nums`, returnÂ *an integer array*Â `counts`Â *where*Â `counts[i]`Â *is the number of smaller elements to the right of*Â `nums[i]`.

# **Example 1:**

# </aside>

def countSmaller(nums):
    def mergeSort(arr, start, end):
        if start >= end:
            return

        mid = (start + end) // 2
        mergeSort(arr, start, mid)
        mergeSort(arr, mid+1, end)
        merge(arr, start, mid, end)

    def merge(arr, start, mid, end):
        temp = []
        i, j = start, mid+1
        count = 0

        while i <= mid and j <= end:
            if nums[i] <= nums[j]:
                temp.append(nums[i])
                counts[i] += count
                i += 1
            else:
                temp.append(nums[j])
                count += 1
                j += 1

        while i <= mid:
            temp.append(nums[i])
            counts[i] += count
            i += 1

        while j <= end:
            temp.append(nums[j])
            j += 1

        nums[start:end+1] = temp

    n = len(nums)
    counts = [0] * n

    mergeSort(nums, 0, n-1)

    return counts
```

```python
# <aside>
# ğŸ’¡ 3. **Sort an Array**

# Given an array of integersÂ `nums`, sort the array in ascending order and return it.

# You must solve the problemÂ **without using any built-in**Â functions inÂ `O(nlog(n))`Â time complexity and with the smallest space complexity possible.

# </aside>
def sortArray(nums):
    def mergeSort(arr, start, end):
        if start >= end:
            return

        mid = (start + end) // 2
        mergeSort(arr, start, mid)
        mergeSort(arr, mid+1, end)
        merge(arr, start, mid, end)

    def merge(arr, start, mid, end):
        temp = []
        i, j = start, mid+1

        while i <= mid and j <= end:
            if nums[i] <= nums[j]:
                temp.append(nums[i])
                i += 1
            else:
                temp.append(nums[j])
                j += 1

        while i <= mid:
            temp.append(nums[i])
            i += 1

        while j <= end:
            temp.append(nums[j])
            j += 1

        nums[start:end+1] = temp

    mergeSort(nums, 0, len(nums)-1)
    return nums

```

```python
# <aside>
# ğŸ’¡ 4. **Move all zeroes to end of array**

# Given an array of random numbers, Push all the zeroâ€™s of a given array to the end of the array. For example, if the given arrays is {1, 9, 8, 4, 0, 0, 2, 7, 0, 6, 0}, it should be changed to {1, 9, 8, 4, 2, 7, 6, 0, 0, 0, 0}. The order of all other elements should be same. Expected time complexity is O(n) and extra space is O(1).

# </aside>

def moveZeroes(nums):
    left = right = 0

    while left < len(nums):
        if nums[left] != 0:
            nums[left], nums[right] = nums[right], nums[left]
            right += 1
        left += 1

    return nums
nums = [1, 9, 8, 4, 0, 0, 2, 7, 0, 6, 0]
print(moveZeroes(nums))  # Output: [1, 9, 8, 4, 2, 7, 6, 0, 0, 0, 0]

```

```python
# <aside>
# ğŸ’¡ 5. **Rearrange array in alternating positive & negative items with O(1) extra space**

# Given anÂ **array of positive**Â andÂ **negative numbers**, arrange them in anÂ **alternate**Â fashion such that every positive number is followed by a negative and vice-versa maintaining theÂ **order of appearance**.Â The number of positive and negative numbers need not be equal. If there are more positive numbers they appear at the end of the array. If there are more negative numbers, they too appear at the end of the array.

# </aside>

def rearrangeArray(nums):
    pos = 0
    neg = 0
    n = len(nums)

    while pos < n and neg < n:
        while pos < n and nums[pos] >= 0:
            pos += 1
        while neg < n and nums[neg] < 0:
            neg += 1

        if pos < n and neg < n:
            nums[pos], nums[neg] = nums[neg], nums[pos]
            pos += 1
            neg += 1

    return nums
nums = [1, -2, 3, -4, 5, -6, 7, -8]
print(rearrangeArray(nums))  # Output: [1, -2, 3, -4, 5, -6, 7, -8]
```

```python
# <aside>
# ğŸ’¡ **6. Merge two sorted arrays**

# Given two sorted arrays, the task is to merge them in a sorted manner.

# </aside>

def mergeArrays(array1, array2):
    merged = []
    i = 0
    j = 0

    while i < len(array1) and j < len(array2):
        if array1[i] <= array2[j]:
            merged.append(array1[i])
            i += 1
        else:
            merged.append(array2[j])
            j += 1

    while i < len(array1):
        merged.append(array1[i])
        i += 1

    while j < len(array2):
        merged.append(array2[j])
        j += 1

    return merged
```

```python
# <aside>
# ğŸ’¡ 7. **Intersection of Two Arrays**

# Given two integer arraysÂ `nums1`Â andÂ `nums2`, returnÂ *an array of their intersection*. Each element in the result must beÂ **unique**Â and you may return the result inÂ **any order**.

# </aside>

def intersection(nums1, nums2):
    intersectionSet = set(nums1)
    intersection = []

    for num in nums2:
        if num in intersectionSet:
            intersection.append(num)
            intersectionSet.remove(num)

    return intersection
```

```python
# <aside>
# ğŸ’¡ 8. **Intersection of Two Arrays II**

# Given two integer arraysÂ `nums1`Â andÂ `nums2`, returnÂ *an array of their intersection*. Each element in the result must appear as many times as it shows in both arrays and you may return the result inÂ **any order**.

# </aside>

def intersect(nums1, nums2):
    countDict = {}
    intersection = []

    for num in nums1:
        countDict[num] = countDict.get(num, 0) + 1

    for num in nums2:
        if num in countDict and countDict[num] > 0:
            intersection.append(num)
            countDict[num] -= 1

    return intersection
nums1 = [1, 2, 2, 1]
nums2 = [2, 2, 3]
print(intersect(nums1, nums2))  # Output: [2, 2]
```
