```python
# 5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
#    Group A: [85, 90, 92, 88, 91]
#    Group B: [82, 88, 90, 86, 87]
#    Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.


# To perform a t-test to determine if there is a significant difference in the mean scores between two groups, we can follow these steps:

# Step 1: Define the null hypothesis (H0) and the alternative hypothesis (Ha):

# H0: There is no significant difference in the mean scores between Group A and Group B.
# Ha: There is a significant difference in the mean scores between Group A and Group B.
# Step 2: Calculate the means of the two groups:

# Group A: [85, 90, 92, 88, 91]
# Mean A = (85 + 90 + 92 + 88 + 91) / 5 = 88.4
# Group B: [82, 88, 90, 86, 87]
# Mean B = (82 + 88 + 90 + 86 + 87) / 5 = 86.6
# Step 3: Calculate the standard deviations of the two groups:

# Group A: [85, 90, 92, 88, 91]
# Standard deviation A = sqrt(((85-88.4)^2 + (90-88.4)^2 + (92-88.4)^2 + (88-88.4)^2 + (91-88.4)^2) / 4) = 2.42
# Group B: [82, 88, 90, 86, 87]
# Standard deviation B = sqrt(((82-86.6)^2 + (88-86.6)^2 + (90-86.6)^2 + (86-86.6)^2 + (87-86.6)^2) / 4) = 2.31
# Step 4: Calculate the t-value using the formula:
# t = (Mean A - Mean B) / sqrt((Standard deviation A^2 / nA) + (Standard deviation B^2 / nB))
# t = (88.4 - 86.6) / sqrt((2.42^2 / 5) + (2.31^2 / 5)) = 0.944

# Step 5: Determine the degrees of freedom (df) using the formula:
# df = nA + nB - 2 = 5 + 5 - 2 = 8

# Step 6: Determine the critical t-value based on the significance level and degrees of freedom. Let's assume a significance level of 0.05 (5%) for a two-tailed test. Looking up the critical t-value from the t-distribution table or using statistical software, we find it to be approximately 2.306.

# Step 7: Compare the calculated t-value with the critical t-value:

# If the calculated t-value is greater than the critical t-value, we reject the null hypothesis (Ha is true), indicating a significant difference in the mean scores between the two groups.
# If the calculated t-value is less than or equal to the critical t-value, we fail to reject the null hypothesis (H0 is true), indicating no significant difference in the mean scores between the two groups.
# In this case, the calculated t-value (0.944) is less than the critical t-value (2.306), so we fail to reject the null hypothesis. Therefore, we conclude that there is no significant difference in the mean scores between Group A and Group B based on the t-test at a significance level of 0.05.





```

```python
# 9. Scenario: A study is conducted to compare the effectiveness of two different medications. The recovery times of the patients in each group are as follows:
#    Medication A: [10, 12, 14, 11, 13]
#    Medication B: [15, 17, 16, 14, 18]
#    Perform an analysis of variance (ANOVA) to determine if there is a significant difference in the mean recovery times between the two medications.


# To perform an analysis of variance (ANOVA) to determine if there is a significant difference in the mean recovery times between the two medications, we can follow these steps:

# Step 1: Calculate the mean of the recovery times for each medication group:

# Medication A: [10, 12, 14, 11, 13]
# Mean of Medication A = (10 + 12 + 14 + 11 + 13) / 5 = 12
# Medication B: [15, 17, 16, 14, 18]
# Mean of Medication B = (15 + 17 + 16 + 14 + 18) / 5 = 16
# Step 2: Calculate the sum of squares within groups (SSW):

# Calculate the sum of squared deviations for each medication group from their respective means:
# Deviations for Medication A = [10 - 12, 12 - 12, 14 - 12, 11 - 12, 13 - 12] = [-2, 0, 2, -1, 1]
# Deviations for Medication B = [15 - 16, 17 - 16, 16 - 16, 14 - 16, 18 - 16] = [-1, 1, 0, -2, 2]
# Calculate the sum of squared deviations for each group:
# SSW for Medication A = (-2^2 + 0^2 + 2^2 + -1^2 + 1^2) = 6
# SSW for Medication B = (-1^2 + 1^2 + 0^2 + -2^2 + 2^2) = 10
# Calculate the total sum of squares within groups (SSW):
# SSW = SSW for Medication A + SSW for Medication B = 6 + 10 = 16
# Step 3: Calculate the sum of squares between groups (SSB):

# Calculate the mean of all the recovery times:
# Mean of all recovery times = (10 + 12 + 14 + 11 + 13 + 15 + 17 + 16 + 14 + 18) / 10 = 14
# Calculate the squared deviations of the means of each group from the mean of all recovery times:
# Deviation for Medication A mean = 12 - 14 = -2
# Deviation for Medication B mean = 16 - 14 = 2
# Calculate the sum of squares between groups:
# SSB = (5 * (-2)^2) + (5 * 2^2) = 20 + 20 = 40
# Step 4: Calculate the degrees of freedom:

# Degrees of freedom between groups (dfB) = Number of groups - 1 = 2 - 1 = 1
# Degrees of freedom within groups (dfW) = Total number of observations - Number of groups = 10 - 2 = 8
# Step 5: Calculate the mean squares:

# Mean squares between groups (MSB) = SSB / dfB = 40 / 1 = 40
# Mean squares within groups (MSW) = SSW / dfW = 16 / 8 = 2
# Step 6: Calculate the F-statistic:

# F-statistic = MSB / MSW = 40 / 2 = 20
# Step 7: Perform an F-test and determine the critical F-value for the chosen significance level (e.g., α = 0.05) and degrees of freedom (dfB and dfW). Compare the calculated F-statistic with the critical F-value.

# Step 8: If the calculated F-statistic is greater than the critical F-value, we reject the null hypothesis and conclude that there is a significant difference in the mean recovery times between the two medications. Otherwise, we fail to reject the null hypothesis.

# In this scenario, the analysis of variance (ANOVA) would be used to determine if there is a significant difference in the mean recovery times between Medication A and Medication B.




```

```python
# 17. Scenario: A study is conducted to compare the performance of three different machine learning algorithms. The accuracy scores for each algorithm are as follows:
#     Algorithm A: [0.85, 0.80, 0.82, 0.87, 0.83]
#     Algorithm B: [0.78, 0.82, 0.84, 0.80, 0.79]
#     Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]
#     Perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms.

    
# To perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms, we need to follow these steps:

# Step 1: Arrange the accuracy scores for each algorithm in a combined dataset:
# Algorithm A: [0.85, 0.80, 0.82, 0.87, 0.83]
# Algorithm B: [0.78, 0.82, 0.84, 0.80, 0.79]
# Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]
# Combined dataset: [0.85, 0.80, 0.82, 0.87, 0.83, 0.78, 0.82, 0.84, 0.80, 0.79, 0.90, 0.88, 0.89, 0.86, 0.87]

# Step 2: Rank the combined dataset in ascending order:
# Combined dataset: [0.78, 0.79, 0.80, 0.80, 0.82, 0.82, 0.83, 0.84, 0.85, 0.86, 0.87, 0.87, 0.88, 0.89, 0.90]
# Ranks: [1, 2, 3.5, 3.5, 5.5, 5.5, 7, 8, 9, 10, 11.5, 11.5, 13, 14, 15]

# Step 3: Calculate the sum of ranks for each algorithm:
# Algorithm A: Sum of ranks = 3.5 + 5.5 = 9
# Algorithm B: Sum of ranks = 1 + 3.5 + 5.5 = 10
# Algorithm C: Sum of ranks = 7 + 8 + 9 + 11.5 + 11.5 + 13 + 14 + 15 = 89

# Step 4: Calculate the mean rank for each algorithm:
# Algorithm A: Mean rank = Sum of ranks / Number of scores in Algorithm A = 9 / 2 = 4.5
# Algorithm B: Mean rank = Sum of ranks / Number of scores in Algorithm B = 10 / 2 = 5
# Algorithm C: Mean rank = Sum of ranks / Number of scores in Algorithm C = 89 / 5 = 17.8

# Step 5: Calculate the test statistic (H):
# H = (12 / (Number of samples * (Number of samples + 1))) * (Σ(Ri - (Number of samples + 1) / 2)² / Ni)
# = (12 / (15 * (15 + 1))) * ((4.5 - 8.0)² / 2 + (5 - 8.0)² / 2 + (17.8 - 8.0)² / 5)
# = (12 / 240) * (3.5² / 2 + 3² / 2 + 9.8² / 5)
# = (0.05) * (6.125 + 4.5 + 19.208)
# = 1.0745

# Step 6: Determine the critical value for the desired significance level (α) and degrees of freedom (df):
# With α = 0.05 and df = Number of samples - 1 = 3 - 1 = 2, the critical value is 5.991 (from the Chi-Square distribution table).

# Step 7: Compare the test statistic (H) with the critical value:
# H (1.0745) < Critical value (5.991)

# Step 8: Make a decision:
# Since the test statistic (H) is less than the critical value, we do not reject the null hypothesis. There is no significant difference in the median accuracy scores between the algorithms.

# Therefore, based on the Kruskal-Wallis test, we conclude that there is no significant difference in the median accuracy scores between Algorithm A, Algorithm B, and Algorithm C.





```

```python
# 18. Scenario: A company wants to analyze the effect of price on sales. The data collected is as follows:
#     Price (in dollars): [10, 15, 12, 8, 14]
#     Sales: [100, 80, 90, 110, 95]
#     Perform a simple linear regression to predict

#  sales based on price.



# To perform a simple linear regression to predict sales based on price, we need to follow these steps:

# Step 1: Organize the data into two variables: price and sales.
# Price (in dollars): [10, 15, 12, 8, 14]
# Sales: [100, 80, 90, 110, 95]

# Step 2: Calculate the mean of price (x̄) and sales (ȳ).
# x̄ = (10 + 15 + 12 + 8 + 14) / 5 = 11.8
# ȳ = (100 + 80 + 90 + 110 + 95) / 5 = 95

# Step 3: Calculate the deviations from the mean for price (x) and sales (y).
# Deviations for price (x):
# x1 = 10 - 11.8 = -1.8
# x2 = 15 - 11.8 = 3.2
# x3 = 12 - 11.8 = 0.2
# x4 = 8 - 11.8 = -3.8
# x5 = 14 - 11.8 = 2.2

# Deviations for sales (y):
# y1 = 100 - 95 = 5
# y2 = 80 - 95 = -15
# y3 = 90 - 95 = -5
# y4 = 110 - 95 = 15
# y5 = 95 - 95 = 0

# Step 4: Calculate the product of the deviations for price and sales (xy).
# xy1 = -1.8 * 5 = -9
# xy2 = 3.2 * -15 = -48
# xy3 = 0.2 * -5 = -1
# xy4 = -3.8 * 15 = -57
# xy5 = 2.2 * 0 = 0

# Step 5: Calculate the squared deviations for price (x²).
# x²1 = (-1.8)² = 3.24
# x²2 = 3.2² = 10.24
# x²3 = 0.2² = 0.04
# x²4 = (-3.8)² = 14.44
# x²5 = 2.2² = 4.84

# Step 6: Calculate the sum of deviations, product of deviations, and squared deviations.
# ∑x = -1.8 + 3.2 + 0.2 - 3.8 + 2.2 = 0.8
# ∑y = 5 - 15 - 5 + 15 + 0 = 0
# ∑xy = -9 - 48 - 1 - 57 + 0 = -115
# ∑x² = 3.24 + 10.24 + 0.04 + 14.44 + 4.84 = 32.8

# Step 7: Calculate the slope (b) of the regression line.
# b = ∑xy / ∑x² = -115 / 32.8 ≈ -3.51

# Step 8: Calculate the intercept (a) of the regression line.
# a = ȳ - b * x̄ = 95 - (-3.51) * 11.8 ≈ 135.10

# Step 9: Write the regression equation.
# The regression equation is: sales = 135.10 - 3.51 * price

# Step 10: Interpret the results.
# The slope (-3.51) indicates that for every one-unit increase in price, the sales decrease by an average of 3.51 units. The intercept (135.10) represents the expected sales when the price is zero.

# Therefore, based on the simple linear regression analysis, the regression equation sales = 135.10 - 3.51 * price can be used to predict sales based on the price.




```
